```python
import pandas as pd
import numpy as np
```

```python
ridgecoef = pd.read_csv("../data/ridge_coef1.csv")
indexmap = pd.read_csv("../data/player_index_map.csv")
contractteam = pd.read_csv("../data/priors_contract+team.csv")
```

# Add names to ridge coef


```python
ridgenames = pd.merge(ridgecoef, indexmap, how="left",left_on="index", right_on = "index")
ridgenames = ridgenames.drop(['Unnamed: 0_x','Unnamed: 0_y'],axis=1)
ridgenames
```

## add coefs to contractteam

```python
finalds = pd.merge(contractteam, ridgenames, how="left",left_on="Name", right_on = "player_name")
#ridgenames = ridgenames.drop(['Unnamed: 0_x','Unnamed: 0_y'],axis=1)
finalds
```

```python
finalds.to_csv(r'../data/finalds.csv')
```
